```python
import pandas as pd
import numpy as np

from sklearn.model_selection import GridSearchCV, cross_val_score
from alg6 import model
```

```python
DATA='ugrin2020-vehiculo-usado-multiclase/'
TRAIN=DATA+'train.csv'
TEST=DATA+'test.csv'

PREPROCESSED_DATA='preprocessed_data/'
RESULTS='results/'
```

```python
# Carga de datos ya procesados
X=np.load(PREPROCESSED_DATA+'binScale-unbalanced.npz')
train = X['arr_0']
label = X['arr_1']
test = X['arr_2']
X.close()
```

```python
#param_grid={'C': [0.5,1,3,5,10,15,20,30,40,50],'kernel':['rbf','poly','sigmoid']}
param_grid={'C': list(range(55,76,5))}
searcher = GridSearchCV(model, param_grid, n_jobs=4,verbose=15,cv=5)
search = searcher.fit(train,label)
```

```python
pd.DataFrame(search.cv_results_)[['param_C','param_kernel','mean_test_score','rank_test_score']].sort_values(by='rank_test_score')
```

```python
print('Mejores parámetros:', search.best_params_)
print('Mejor puntuación:', search.best_score_)
model.set_params(**(search.best_params_))
```

```python
scores=cross_val_score(model, train, label)
print(scores)
print(np.mean(scores))
```
