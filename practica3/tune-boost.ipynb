```python
import pandas as pd
import numpy as np

from sklearn.model_selection import GridSearchCV, cross_val_score
from alg7 import model
```

```python
DATA='ugrin2020-vehiculo-usado-multiclase/'
TRAIN=DATA+'train.csv'
TEST=DATA+'test.csv'

PREPROCESSED_DATA='preprocessed_data/'
RESULTS='results/'
```

```python
# Carga de datos ya procesados
X=np.load(PREPROCESSED_DATA+'binScale.npz')
train = X['arr_0']
label = X['arr_1']
test = X['arr_2']
X.close()
```

```python
train.shape
```

```python
param_grid={'n_estimators':[400,450,500,550], 'learning_rate':[0.1,0.11,1.2,1.3,1.4]}
searcher = GridSearchCV(model, param_grid, n_jobs=4,verbose=15,cv=3)
search = searcher.fit(train,label)
```

```python
pd.DataFrame(search.cv_results_)[['param_n_estimators','param_learning_rate','mean_test_score','rank_test_score']].sort_values(by='rank_test_score')
```

```python
print('Mejores parámetros:', search.best_params_)
print('Mejor puntuación:', search.best_score_)
model.set_params(**(search.best_params_))
```

```python
scores=cross_val_score(model, train, label)
print(scores)
print(np.mean(scores))
```
